```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
```

```python
def calc_PIC_z(system_name, N_water):

    colnames = []
    for g in np.arange(1.5,5.1,0.1):
        colnames.append('r_'+str(g))
    PFASframes = []                        # each element of the list contains the dataframe for a individual simulation replicate
    waterframes = []
    for i in range(1,7):
        PFASframes.append(pd.read_csv(str(i)+system_name+'tstp_nPFAS_r.txt',delimiter=" ",names=colnames))
        waterframes.append(pd.read_csv(str(i)+system_name+'tstp_WATER_r.txt',delimiter=" ",names=colnames))

    N_PFA = 100
    PIC_r_80_lst = []

    for k in range(6):    #index of the duplicate
        for i in [799]:
            P_r_80 = PFASframes[k].iloc[i,:].to_numpy() #this is an array of how the PIC varies as a function of r at the timestep i 
            W_r_80 = waterframes[k].iloc[i,:].to_numpy()
            P_b_80 = N_PFA - P_r_80
            W_b_80 = N_water - W_r_80
            PIC_r_80 = P_r_80 - (W_r_80*(P_b_80/W_b_80))
        PIC_r_80_lst.append(PIC_r_80)
    PIC_r_80_arr = np.array(PIC_r_80_lst)
    mean_PIC_r=np.mean(PIC_r_80_arr, axis=0)
    np.savetxt(system_name+'mean_PIC(r).txt', mean_PIC_r)
    std_PIC_r=np.std(PIC_r_80_arr, axis=0)
    np.savetxt(system_name+'std_PIC(r).txt', std_PIC_r)
    stderror_PIC_z= stats.sem(PIC_r_80_arr, axis=0, ddof=1)
    np.savetxt(system_name+'ste_PIC(z).txt', stderror_PIC_z)
```

```python
import matplotlib.cm as cm
from matplotlib.colors import ListedColormap
import matplotlib.colors as mcolors
import colorsys

def adjust_lightness(cmap, factor):
    new_colors = []
    for i in range(cmap.N):
        r, g, b, a = cmap(i)
        h, l, s = colorsys.rgb_to_hls(r, g, b)
        l = max(0, min(1, l * factor))
        r_new, g_new, b_new = colorsys.hls_to_rgb(h, l, s)
        new_colors.append((r_new, g_new, b_new, a))
    return ListedColormap(new_colors, name=f"{cmap.name}_light{factor}") 

levels = np.linspace(0,1,20)
custom_cmap = mcolors.LinearSegmentedColormap.from_list( "pink_white_blue", ["blue","white", "deeppink"], N=len(levels) - 1)
lighter_cmap = adjust_lightness(custom_cmap, factor=1.1)
```
